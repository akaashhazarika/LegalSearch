```python
import pandas as pd

from numpy import array
from keras.preprocessing.text import one_hot
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Conv1D, GlobalMaxPool1D, LSTM
from keras.layers import Dense, Dropout, MaxPooling1D
from keras.layers import Flatten
from keras.layers.embeddings import Embedding

data = pd.read_csv('LABEL.txt', sep = '\t', header=None, encoding = "ISO-8859-1")

print(data)
```

```python
from sklearn.model_selection import train_test_split

vocab_size = 5000

encoded_docs = [one_hot(d, vocab_size) for d in data[1]]
print(encoded_docs)

max_length = len(max(data[1])) + 1

print(max_length)

# x and y
padded_docs = pad_sequences(encoded_docs, maxlen=max_length, padding='post')
y_data = data[0]
print(padded_docs)
print(y_data)

X_train, X_test, y_train, y_test = train_test_split( padded_docs, y_data, test_size=0.2, random_state=42)
```

```python
model = Sequential()
model.add(Embedding(vocab_size, 64, input_length=max_length))
model.add(LSTM(128))
model.add(Dense(64, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adadelta', loss='binary_crossentropy', metrics=['accuracy'])
print(model.summary())
```

```python
#fit
model.fit(X_train, y_train, batch_size=512, epochs = 300, 
    verbose=1, validation_data= (X_test, y_test))
```

```python
predictions = model.predict(X_test)
print(predictions)

for i in range(len(predictions)):
    if predictions[i]>0.5:
        predictions[i] = 1
    else:
        predictions[i] = 0

print(predictions)
```

```python
from sklearn.metrics import confusion_matrix, classification_report


print(classification_report(y_test, predictions))
confusion_matrix(y_test, predictions)

```
